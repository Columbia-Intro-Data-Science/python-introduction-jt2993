# Introduction


In this homework, you'll be required to load in a dataset which has about 500 features. By using
Lasso ($L^1$) regression, we'll find the optimal constraint on the $L^1$ norm which gives us the best
$R^2$. Then we'll plot the results.

Recall we minimize the following on ** training data: $(x_i,y_i)$**

$$\min_{\beta} \frac{1}{N} \sum_{i=1}^N (y_i - \beta \cdot x_i)^2 + \lambda \|\beta \|_{L^1}.$$


Denoting $\beta_{\lambda}$ as the minimum of the above, we then choose $\lambda$ to maximize $R^2$ on **testing data: $(x_j,y_j)$**

$$ \max_{\lambda} 1 - \frac{\sum_{j} (y_j - \beta_{\lambda} \cdot x_j)^2}{\sum_j (y_j - \bar y)^2}$$

a) Load in hw2data.csv from ../data into a pandas dataframe.

```python
import pandas as pd
import pylab as plt
import seaborn
from sklearn.linear_model import LinearRegression
import numpy.random as nprnd
import random
import json
import matplotlib.pyplot as plt
import numpy as np

pd.set_option('display.max_columns', 500)
%matplotlib inline

df = pd.read_csv('https://raw.githubusercontent.com/doriang102/APM4990-/master/data/hw2data.csv')
df.head()

```

b) Set y to be the y variable in the dataframe from a and X to be the remaining features.

```python
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler

y = df['y']
X = df.drop(['y'], 1)

scaler = preprocessing.StandardScaler()
X_scaled = X.as_matrix().astype(np.float)
X_scaled = scaler.fit_transform(X)
```

c) As shown in the Booking.com example, using Lasso regression, find the regularization strength
which optimizes the $R^2$. 

**Hint:** Take a range of alpha from `np.logspace(-8,-3,1000)`

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split

#Creating the training and testing set
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size = 0.20, random_state = 42)

alphas = np.logspace(-8,-3,1000)
train_errors = []
test_errors = []
scores = []
for alpha in alphas:
    regr = Lasso(alpha = alpha)
    # Then, I train the model using training sets
    regr.fit(X_train, y_train)
    train_errors.append(regr.score(X_train, y_train))
    scores.append(regr.score(X_test, y_test))

plt.plot(alphas,scores)
```

```python
# To find the regularisation strength which optimises R^2,
alpha_optim = alphas[np.argmax(scores)]
alpha_optim
```

d) Plot the training perforamnce versus the testing performance, and observe whree the test performance is
maximized. I've written an outline of the code you need.

```python
test_errors = scores

plt.semilogx(alphas, train_errors, label = 'Train')
plt.semilogx(alphas, test_errors, label = 'Test')
plt.vlines(alpha_optim, plt.ylim()[0], np.max(test_errors), color='k',
           linewidth=3, label='Optimum on test')
plt.legend(loc='lower left')
plt.ylim([0, 1.2])
plt.xlabel('Regularization parameter')
plt.ylabel('Performance')

plt.show()
```

e) Plot the top coefficients based on this optimal paramter. Why do you think so many are zero? 

```python
regr = Lasso(alpha = alpha_optim)
regr.fit(X_train, y_train)
regr.coef_
```

```python
data = pd.DataFrame()
data['name'] = np.arange(len(regr.coef_))
data['value'] = regr.coef_
data = data.sort_values('value')

data[0:30].plot(x = 'name', y = 'value', kind = 'bar', figsize=(12,10))
```

So many coefficients for lasso are zero because L1 shrinkage which is why there are corners in a constraint. The cefficient corresponding to the axis will be shrunk to zero if the sum of squares "hits'' one of these corners. As the number of dimensions increases, there will be more number of corners which result in so many coefficients being zero.

f) Compute the $R^2$ with the optimal coefficient found above on 5 folds using cross_val_score and plot the
results. Does the model work well on all random subsets?

```python
from sklearn.model_selection import cross_val_score

regr = Lasso(alpha = alpha_optim)
scores = cross_val_score(regr, X_scaled, y, cv=5)
plt.ylim([-1,1])
plt.xlabel('lambda')
plt.ylabel('R^2')
plt.title('Performance on 5 folds with lambda=' + str(alpha))
plt.bar(range(1,6),scores)
plt.show()
```

From the graph above, since the R^2 value is almost 1.00 for the different subsets, this show that this model does fit the data pretty well and would be able to work well on all the random subsets.

f) Repeat e) but using cross validation. Use error bars on the features which are the standard deviation of the 
coefficiens obtained above. For this problem I"ll walk you through the code. You just need to apply your optimal
$\alpha$ found above. 

```python
from sklearn.cross_validation import KFold
from sklearn import preprocessing
def run_cv_coeffs(X,y,clf_class,**kwargs):
    # Construct a kfolds object
    kf = KFold(len(y),n_folds=5,shuffle=True)
    y_pred = y.copy()
    coeffs=[]
    # Iterate through folds
    for train_index, test_index in kf:
        
        X_train, X_test = X[train_index], X[test_index]
        y_train = y[train_index]
        # Initialize a classifier with key word arguments
        clf = clf_class(**kwargs)
        clf.fit(X_train,y_train)
        y_pred[test_index] = clf.predict(X_test)
        coeffs.append(clf.coef_)
    return coeffs



scaler = preprocessing.StandardScaler()
X_scaled = X.as_matrix().astype(np.float)
X_scaled = scaler.fit_transform(X)

coeffs=run_cv_coeffs(X_scaled,np.array(y),Lasso,alpha=alpha_optim)
```

```python
X_Scaled = pd.DataFrame(X_scaled)
def get_coeffs(coeffs):
    coeffs_avgd = [(coeffs[0][i] + coeffs[1][i] + coeffs[2][i] + coeffs[3][i] + coeffs[4][i])/5 for i in range(0,len(X_Scaled.columns))]
    coeffs_std = [np.std([coeffs[0][i],coeffs[1][i],coeffs[2][i],coeffs[3][i],coeffs[4][i]]) for i in range(0,len(X_Scaled.columns))]
    return coeffs_avgd, coeffs_std
coeffs_avg,coeffs_std=get_coeffs(coeffs)
```

```python
dfCoeffs = pd.DataFrame({'type':X_Scaled.columns.values, 'coef':coeffs_avg, 'std':coeffs_std})
dfCoeffs = dfCoeffs[(dfCoeffs['coef']>1) |(dfCoeffs['coef']<-1) ]
plt.figure(figsize=(15,15))
dfCoeffs_sorted = dfCoeffs.sort_values(['coef'])[::-1]
yerr_vals = dfCoeffs_sorted['std'].values
dfCoeffs_sorted.plot(x='type',y='coef',kind='bar',yerr=yerr_vals,figsize=(15,15))
```
